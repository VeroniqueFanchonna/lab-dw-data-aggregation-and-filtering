```python
import pandas as pd

url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis.csv"
df = pd.read_csv(url)
# local save for future use
df.to_csv("marketing_customer_analysis.csv", index=False)
```

```python
#quick cleaning of columns and cases + display first 10 rows
df.columns = [col.lower().replace(' ', '_') for col in df.columns]
df.head(10)
```

```python
#filtring out only response "Yes"
df_yes = df[df['response'] == 'Yes']
#aggregating insight by mean claim amount by education and gender
result = df_yes.groupby(['education', 'gender'])[['total_claim_amount']].mean().reset_index()
result
```

```python
"""
2. Using the original Dataframe, analyze:
   - the average `monthly_premium` and/or customer lifetime value by `policy_type` and `gender` for customers who responded "Yes", and
   - compare these insights to `total_claim_amount` patterns, and discuss which segments appear most profitable or low-risk for the company.
"""
# Global claims coast comparaired to monthly_premium for "Yes" response
comparison = df_yes.groupby(['policy_type', 'gender'])[['customer_lifetime_value', 'total_claim_amount']].mean().reset_index()

# New column  "Profitability Ratio" in order to compare (income / coast)
comparison['profit_ratio'] = comparison['customer_lifetime_value'] / comparison['total_claim_amount']
comparison.sort_values(by='profit_ratio', ascending=False)
```

```python
# 1. Global claims coast per Policy, Gender  to monthly_premium for "Yes"
comparison = df_yes.groupby(['policy_type', 'gender']).agg({
    'customer_lifetime_value': 'mean',
    'total_claim_amount': 'mean',
    'monthly_premium_auto': 'mean'
}).reset_index()

# 2. New column  "Profitability Ratio" in order to compare (income / coast)
comparison['profit_ratio'] = comparison['customer_lifetime_value'] / comparison['total_claim_amount']

# 3. Tri pour voir le segment le plus "Safe" (Profit Ã©levÃ© + Claim bas)
analysis_ready = comparison.sort_values(by=['profit_ratio', 'total_claim_amount'], ascending=[False, True])
analysis_ready
```

```python
"""The most profitable segments are those with the highest Profit Ratio and the lowest at the top of the table : corporation auto."""
```

```python
# Step A : Compute the total number of customers per state
state_analysis = df.groupby('state')['customer'].count().reset_index()

# Ã‰tape B : Filtrerint for market over 500
large_markets = state_analysis[state_analysis['customer'] > 500]

print(large_markets)
```

```python
# We use the original DataFrame as requested
# We group by Education and Gender, then we calculate max, min, and median
clv_stats = df.groupby(['education', 'gender'])['customer_lifetime_value'].agg(['max', 'min', 'median']).reset_index()

# We rename the columns so that it is more professional
clv_stats.columns = ['Education', 'Gender', 'Max CLV', 'Min CLV', 'Median CLV']

clv_stats
```

```python
# 5. Tableau policies vendues par Ã©tat (lignes) et mois (colonnes)
df['effective_to_date'] = pd.to_datetime(df['effective_to_date'])
df['month'] = df['effective_to_date'].dt.month

pivot_month_state = df.pivot_table(
    index='state', 
    columns='month', 
    values='customer', 
    aggfunc='count', 
    fill_value=0
).round(0).astype(int)

print("Policies par Ã©tat et mois :")
pivot_month_state
```

```python
# 6. Top 3 Ã©tats avec policies par mois
# Ã‰tape 1 : Compter policies par Ã©tat + mois, trier par total descending
policies_by_state_month = df.groupby(['state', 'month']).size().reset_index(name='policies_sold')
top_states_total = policies_by_state_month.groupby('state')['policies_sold'].sum().sort_values(ascending=False).head(3).index

# Ã‰tape 2 : Filtrer top 3 Ã©tats et pivoter
top3_pivot = policies_by_state_month[
    policies_by_state_month['state'].isin(top_states_total)
].pivot_table(
    index='state', 
    columns='month', 
    values='policies_sold', 
    fill_value=0
).round(0).astype(int)

print("Top 3 Ã©tats - Policies par mois :")
top3_pivot
```
